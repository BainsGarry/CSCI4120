```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from yellowbrick.cluster import KElbowVisualizer
from sklearn.metrics import confusion_matrix
import numpy as np

# Generate sample data
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Determine the best K using the Elbow Method
visualizer = KElbowVisualizer(KMeans(), k=(2, 10))
visualizer.fit(X)
visualizer.show()

# Get the optimal K from the visualizer
best_k = visualizer.elbow_value_

# Perform K-Means clustering with the best K value
kmeans = KMeans(n_clusters=best_k, random_state=0).fit(X)
y_predict = kmeans.predict(X)

# Calculate accuracy
accuracy = np.mean(y_true == y_predict)
print("Best K:", best_k)
print("Accuracy: " + str(round(accuracy, 2)))

# Construct and print the confusion matrix
conf_matrix = confusion_matrix(y_true, y_pred)
print('Confusion Matrix:')
print(conf_matrix)


```
